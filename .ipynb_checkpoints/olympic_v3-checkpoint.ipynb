```python
import pandas as pd
import heapq


df = pd.read_json('olympic.json')

df
```

```python
# separating the json file with the scraped data into 3 parts
# this first part consists of the overall information about all the Winter Olympics that occured
df1 = df.drop(['disciplines_details','medals_per_country'],axis=1)

# this consist of the medals statistic
df2 = df[['medals_per_country', 'event_title']]

# this consist of specific event detail
df3 = df[['disciplines_details', 'event_title']]
```

```python
# creating 2 new empty columns
df1.insert(0,'event_number',0)
df1.insert(1,'event_year',0)
```

```python
# stripping the 'event_title' into 2 strings and allocating them in the corresponding new columns

for i in range(df1.shape[0]):
    event_title_stripped = df1['event_title'][i].split('/',1)
    df1['event_number'][i] = event_title_stripped[0]
    df1['event_year'][i] = event_title_stripped[1]
```

```python
# creating 2 new empty columns
df1.insert(2,'event_city',0)
df1.insert(3,'event_country',0)
```

```python
# stripping the 'event_title' into 2 strings and allocating them in the corresponding new columns

for i in range(df1.shape[0]):
    event_place_stripped = df1['event_place'][i].split(', ',1)
    df1['event_city'][i] = event_place_stripped[0]
    df1['event_country'][i] = event_place_stripped[1]
```

```python
df1
```

```python
df1.drop(['event_place','event_title'],axis=1, inplace=True)
```

```python
df1.sort_values(by=['event_year'], inplace=True)
```

```python
df1.set_index('event_number', inplace=True)
```

```python
# removing unwanted text in the 'n_countries' column
for i in range(df1.shape[0]):
    df1['n_countries'][i] = df1['n_countries'][i].replace(' countries', '')
    
# converting strings into integers
df1['n_countries'] = pd.to_numeric(df1['n_countries'])
```

```python
df1.head()
```

```python
metric_list = ['n_participants','n_countries','n_medals','n_disciplines']

mean_list = []

for element in metric_list:
    mean = int(round(df1[element].mean(),0))
    mean_list.append(mean)

print(f'Through all the Winter Olympic event, the average number of \n participants = {mean_list[0]} \n countries = {mean_list[1]} \n medals = {mean_list[2]} \n disciplines = {mean_list[3]}')
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

x = df1['event_year']
y = df1['n_participants']
z = df1['n_disciplines']

ax.set_title('Evolution of number of participants and disciplines per year')

ax.bar(x,y, label='Number of participants')
ax.legend(loc=9)


ax2=ax.twinx()
ax2.plot(x, z,'Hr', markersize=8, label='number of disciplines')

ax.set_xlabel('event_year')
ax.set_ylabel('n_participants')
ax2.set_ylabel('n_disciplines')

plt.annotate("World War II **", xy=(1942, 8), xytext=(1934, 12),arrowprops=dict(arrowstyle='fancy', facecolor='black'),
             fontstyle='italic', fontsize=12)

plt.annotate("Separation between Summer and Winter Olympics*", xy=(1992, 12), xytext=(1922, 14),arrowprops=dict(arrowstyle='fancy', facecolor='black'),
             fontstyle='italic', fontsize=12)
plt.legend()

plt.show()
```

*The 1992 Winter Olympic Games (Albertville, France) are the last ones to be happening the same year as the Summer Olympic Games.
Hence the 1994 Winter Olympic Games (Lillehammer, Norway) are the first occuring separately than the Summer Olympic Games.

** The 1940 Olympic Games were attributed ti Sapporo, Japan. These Games were canceled in 1938 due to the japanase invasion in China.

```python
# helper function to sort value by year
def sort_value(df):
    df.insert(1,'event_year',0)
    for i in range(df2.shape[0]):
        event_title_stripped = df['event_title'][i].split('/',1)
        df['event_year'][i] = event_title_stripped[1]
    df.drop(columns=['event_title'], inplace=True)
    df.sort_values(by=['event_year'], inplace=True)

    # reset index
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
def medal_list(df):
    """
    function take in a dataframe
    output: list of number of medals per category
    """
    # a df where columns name is every country that has won a medal
    test_df = pd.json_normalize(df["medals_per_country"], max_level=0)
    n_rows = len(df)
    count = 0 # initiate count of rows
    total = 0 # initiate count of total medal
    gold = 0
    silver = 0
    bronze = 0
    total_list = []
    gold_list = []
    silver_list = []
    bronze_list = []
    for column in test_df:
        # iterate through row
        for index, row in df.iterrows():
            # we know that there are n_rows in a column. when we run through n_row, we add value for next column
            if count <= n_rows:
                count += 1
                try:
                    total += int(row['medals_per_country'][column]['total_medal'])
                except:
                    continue
                try:
                    gold += int(row['medals_per_country'][column]['gold_medal'])
                except:
                    continue
                try:
                    silver += int(row['medals_per_country'][column]['silver_medal'])
                except:
                    continue
                try:
                    bronze += int(row['medals_per_country'][column]['bronze_medal'])
                except:
                    continue
            else:
                count = 0
                total_list.append(total)
                gold_list.append(gold)
                silver_list.append(silver)
                bronze_list.append(bronze)

                total = 0
                gold = 0
                silver = 0
                bronze = 0
    return total_list, gold_list, silver_list, bronze_list, test_df
```

```python
total_list, gold_list, silver_list, bronze_list, test_df = medal_list(df)
len(total_list)
```
