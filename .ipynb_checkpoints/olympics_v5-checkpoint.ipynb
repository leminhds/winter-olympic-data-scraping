```python
import sqlite3
import sys
import pandas as pd
from pandas.core.frame import DataFrame
import heapq
import matplotlib.pyplot as plt
import numpy as np
```

```python
DATABASE_PATH = 'olympic2.db'
```

```python
def connect_database():
    connection = sqlite3.connect(DATABASE_PATH)
    cursor = connection.cursor()
    
    connection.commit()
    
    return connection, cursor
```

```python
connection, cursor = connect_database()
```

```python
df1 = pd.read_sql_query("""
    SELECT * FROM events
""", connection)
```

```python
df1
```

## Analysis of facts per edition

```python
# creating 2 new empty columns
df1.insert(0,'event_number',0)
df1.insert(1,'event_year',0)
```

```python
# stripping the 'event_title' into 2 strings and allocating them in the corresponding new columns

for i in range(df1.shape[0]):
    event_title_stripped = df1['event_title'][i].split('/',1)
    df1['event_number'][i] = event_title_stripped[0]
    df1['event_year'][i] = event_title_stripped[1]
```

```python
# creating 2 new empty columns
df1.insert(2,'event_city',0)
df1.insert(3,'event_country',0)
```

```python
# stripping the 'event_place' into 2 strings and allocating them in the corresponding new columns

for i in range(df1.shape[0]):
    event_place_stripped = df1['event_place'][i].split(', ',1)
    df1['event_city'][i] = event_place_stripped[0]
    df1['event_country'][i] = event_place_stripped[1]
```

```python
df1.drop(['event_place','event_title'],axis=1, inplace=True)
```

```python
df1.sort_values(by=['event_year'], inplace=True)
```

```python
df1.set_index('event_number', inplace=True)
```

```python
# removing unwanted text in the 'n_countries' column
for i in range(df1.shape[0]):
    df1['n_countries'][i] = df1['n_countries'][i].replace(' countries', '')
    
# converting strings into integers
df1['n_countries'] = pd.to_numeric(df1['n_countries'])
```

```python
df1
```

```python
metric_list = ['n_participants','n_countries','n_medals','n_disciplines']

mean_list = []
evolution_list = []

for element in metric_list:
    mean = int(round(df1[element].mean(),0))
    mean_list.append(mean)
    
for element in metric_list:
    evolution = int(round((df1[element][(df1.shape[0]) - 1] / df1[element][0]) * 100,2))
    evolution_list.append(evolution)

print(f'''Through all the Winter Olympic events, the average number and the evolution between the first and last edition, respectively, 
of: \n participants = {mean_list[0]} & {evolution_list[0]} %
\n countries = {mean_list[1]} & {evolution_list[1]} %
\n medals = {mean_list[2]} & {evolution_list[2]} %
\n disciplines = {mean_list[3]} & {evolution_list[3]} %''')
```

The increase in the number of participants and countries represented is particularly relevant, considering this is in less than 100 years. This demonstrates the development of sport in the world, with the mediatisation that accompanies it: Winter Olympics are the 5th most popular sport events in the world (2billion viewers in the last edition).

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

x = df1['event_year']
y = df1['n_participants']
z = df1['n_disciplines']

ax.set_title('Evolution of the number of participants and disciplines per year')

ax.bar(x,y, label='Number of participants')
ax.legend(loc=9)


ax2=ax.twinx()
ax2.plot(x, z,'Hr', markersize=8, label='number of disciplines')

ax.set_xlabel('event_year')
ax.set_ylabel('n_participants')
ax2.set_ylabel('n_disciplines')

plt.annotate("World War II **", xy=(1942, 8), xytext=(1934, 12),arrowprops=dict(arrowstyle='fancy', facecolor='black'),
             fontstyle='italic', fontsize=12)

plt.annotate("Separation between Summer and Winter Olympics*", xy=(1992, 12), xytext=(1922, 14),arrowprops=dict(arrowstyle='fancy', facecolor='black'),
             fontstyle='italic', fontsize=12)
plt.legend()

plt.show()
```

*The 1992 Winter Olympic Games (Albertville, France) are the last ones to be happening the same year as the Summer Olympic Games.
Hence the 1994 Winter Olympic Games (Lillehammer, Norway) are the first occuring separately than the Summer Olympic Games.

** The 1940 Olympic Games were attributed ti Sapporo, Japan. These Games were canceled in 1938 due to the japanase invasion in China.

```python
# stripping the 'event_title' into 2 strings and allocating them in the corresponding new columns

for i in range(df2.shape[0]):
    event_title_stripped = df2['event_title'][i].split('/',1)
    df2['event_number'][i] = event_title_stripped[0]
    df2['event_year'][i] = event_title_stripped[1]
```

```python
df2
```

```python
gold_s = df2.groupby('country')['gold'].sum()
silver_s = df2.groupby('country')['silver'].sum()
bronze_s = df2.groupby('country')['bronze'].sum()

gold_list = list(gold_s)
silver_list = list(silver_s)
bronze_list = list(bronze_s)

gold_country_list = list(gold_s.index)
silver_country_list = list(silver_s.index)
bronze_country_list = list(bronze_s.index)
```

```python
medal_lists = [gold_list, silver_list, bronze_list]
country_list = [gold_country_list, silver_country_list, bronze_country_list]
```

Below is one of the code previously used to create the pie charts.
It was based on the previous dataframes produced with the JSON files.
A simpler and quicker version was found later.

```python
# defining the numbers of the best performing countries we want to analyse
top = 7

medal_list_name = ['total','gold','silver','bronze']

for i in range(3):
    
    # getting the list for the 4 metric from the function
    medallist1 = medal_lists[i]
    # finding the index
    index = heapq.nlargest(top, range(len(medallist1)), key=medallist1.__getitem__)
    # getting the value corresponding to the index
    index_value = heapq.nlargest(top, medallist1)
    # returning the corresponding country name of the index calculated
    
    topcountries = []
    for j in index:
        topcountry = country_list[i][j]
        topcountries.append(topcountry)
    # removing from the list the value of the top x and summing them
    other_countries_sum = sum(list(set(medallist1) - set(index_value)))
    # adding the value of 'Others' to the list
    index_value.append(other_countries_sum)
    # appending the countries list wiht 'Others', its corresponding value being 'other_countries_sum'
    topcountries.append('Others')
        
    def func(pct, allvalues):
        absolute = int(pct / 100.*np.sum(allvalues))
        return "{:.1f}%".format(pct, absolute)

    fig, ax = plt.subplots(figsize =(10, 8))

    wedges, texts, autotexts = ax.pie(index_value, autopct = lambda pct: func(pct, index_value), labels = topcountries, startangle = 90)

    ax.set_title(f"Percentage of all {medal_list_name[i]} medals won in all Winter Olympics editions combined")
    
    plt.show()
```

```python
gold_df = DataFrame({'total_gold' : df2.groupby('country')['gold'].sum()}).reset_index()
```

```python
# function that creates a pie chart, for a different medal type (total, gold, silver or bronze)
# for a defined number of top countries

def create_pie(df, medal_type,top=10):
    new_df = df.copy()
    test_df = new_df.replace(new_df.groupby('country')[medal_type].sum().sort_values(ascending=False).index[top:], 'Other').groupby('country')[medal_type].sum()
    medal_df = DataFrame({medal_type : test_df}).reset_index() 
    
    # functions that calculates the percentage
    def func(pct, allvalues):
        absolute = int(pct / 100.*np.sum(allvalues))
        return "{:.1f}%".format(pct, absolute)

    fig, ax = plt.subplots(figsize =(10, 8))

    wedges, texts, autotexts = ax.pie(medal_df[medal_type], autopct = lambda pct: func(pct, medal_df[medal_type]), labels = medal_df['country'], startangle = 90)

    ax.set_title(f"Percentage of {medal_type} medals won in all Winter Olympics editions combined")    
```

```python
create_pie(df2,'silver')
```

```python
# function that lists the countries that appeared 'n_appearance'th times in the Winter Olympics
def country_appearance(n_appearance):
    
    # list of unique country participants 
    country_list = list(df2['country'].unique())
    
    if n_appearance == 23:
        print(f'These countries were present in all Winter Olympics editions:')
     
    elif n_appearance < 23:
        print(f'The countries that are present in at least {n_appearance} Winter Olympics editions are:')
    
    for elem in country_list:
        
        # counting the appearances
        if (df2[df2['country'] == elem].shape[0]) == n_appearance:
            print(elem)

        elif (df2[df2['country'] == elem].shape[0]) >= n_appearance:
            print(elem)         
```

```python
country_appearance_list = []


# looping to create a list of total apperances per country
for elem in list((df2.groupby('country')['total'].sum()).index):
    appearance = (df2[df2['country'] == elem].shape[0])
    country_appearance_list.append(appearance)   

x = country_appearance_list
y = list(df2.groupby('country')['total'].sum())

plt.scatter(x,y)

plt.xlabel('number of apperances')
plt.ylabel('total medals won')

plt.show()
```
